### Anáisis de los datos

Para el análisis de los datos, utilizaremos los datasets anteriormente generados a partir de los datos obtenidos en el Servicio de datos abiertos del Ajuntament de Barcelona: https://opendata-ajuntament.barcelona.cat/es/, concrtamente de los datasets:

-	Padrón de vehículos. Antigüedad por tipo de vehículo del parque de vehículos de la ciudad de Barcelona.
-	Tipología del parque de vehículos de la ciudad de Barcelona según distintivo ambiental.

Los DataSets que analizaremos y que hemos generados anteriormente son:

- Antigüedad_tipo_TOTAL.csv: Antigüedad por tipo de vehículo del parque de vehículos de la ciudad de Barcelona - Años 2019, 2020 y 2021

### Análisis preliminar

- Determinamos el total de vehículos durante los años 2019, 2020 y 2021 en la ciudad de Barcelona en función de:

    - su antigüedad
    - su distintivo ambiental (etiqueta)
    
- Con la visualización gráfica podemos observar la evolución del número de vehiculos por antigüedad y distintivo ambiental durante los años 2019, 2020 y 2021.

Análisis preliminar por __Antigüedad__:

```python
# Agrupamos los datos y calculamos el total de vehículos
Antigüedad_grouped_df = Antigüedad_tipo_TOTAL.groupby(['Antiguitat', 'Any'])['Nombre'].sum().reset_index()

# Tamaño del gráfico
plt.figure(figsize=(16, 8))

# Gráfico de barras
Antigüedad_bar = sns.barplot(data=Antigüedad_grouped_df, x='Antiguitat', y='Nombre', hue='Any')

# Rotación de las etiquetas del eje x
Antigüedad_bar.set_xticklabels(Antigüedad_bar.get_xticklabels(), rotation=90)

# Etiquetas a las barras
for p in Antigüedad_bar.patches:
    Antigüedad_bar.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'bottom', rotation=90, xytext = (0, 5), textcoords = 'offset points')
       
# Título y etiquetas de los ejes
plt.title('Total de Vehículos por Antigüedad y Año')
plt.xlabel('Antigüedad')
plt.ylabel('Total Vehículos')

plt.show()
```

Análisis preliminar por __Distintivo Ambiental__:

```python
# Agrupamos los datos y calculamos el total de vehículos
Distintivo_AMB_grouped_df = Distintivo_AMB_TOTAL.groupby(['Desc_Distintiu_Ambiental', 'Any'])['Nombre'].sum().reset_index()

# Tamaño del gráfico
plt.figure(figsize=(12, 8))

# Gráfico de barras
Distintivo_AMB_bar = sns.barplot(data=Distintivo_AMB_grouped_df, x='Desc_Distintiu_Ambiental', y='Nombre', hue='Any')

# Etiquetas a las barras
for p in Distintivo_AMB_bar.patches:
   Distintivo_AMB_bar.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

# Título y etiquetas de los ejes
plt.title('Total de Vehículos por Distintivo Ambiental y Año')
plt.xlabel('Distintivo Ambiental')
plt.ylabel('Total Vehículos')

plt.show()
```

Análisis preliminar por __Tipo__:

```python
# Agrupamos los datos y calculamos el total de vehículos
Distintivo_AMB_grouped_df = Distintivo_AMB_TOTAL.groupby(['Tipus_Vehicles', 'Any'])['Nombre'].sum().reset_index()

# Tamaño del gráfico
plt.figure(figsize=(12, 8))

# Gráfico de barras
Distintivo_AMB_bar = sns.barplot(data=Distintivo_AMB_grouped_df, x='Tipus_Vehicles', y='Nombre', hue='Any')

# Etiquetas a las barras
for p in Distintivo_AMB_bar.patches:
   Distintivo_AMB_bar.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'bottom', rotation=90, xytext = (0, 5), textcoords = 'offset points')

# Título y etiquetas de los ejes
plt.title('Total de Vehículos por Tipo y Año')
plt.xlabel('Tipo de vehículo')
plt.ylabel('Total Vehículos')

plt.show()
```

### Análisis distribución vehículos por distritos

- Analizamos en que proporción se distribuyen los vehículos por distritos durante los años 2019, 2020 y 2021 en función de:

    - por tipo
    - su antigüedad
    - su distintivo ambiental (etiqueta)
    
- Con la visualización gráfica podemos observar que proporción vehículos en función su tipo, antigüedad y distintivo ambiental tiene cada distrito.

Proporcion de vehículos en cada uno de los Distritos por __tipo__ y __Año__:

```python
# Filtramos los datos por años (2019, 2020 y 2021):
Antigüedad_filtered_df = Antigüedad_tipo_TOTAL[Antigüedad_tipo_TOTAL['Any'].isin([2019, 2020, 2021])]

# Lista de valores únicos de 'Nom_Districte'
Distritos = Antigüedad_filtered_df['Nom_Districte'].unique()

# Tamaño y diseño de los subplots
fig, axes = plt.subplots(nrows=len(Distritos), ncols=3, figsize=(15, 5*len(Distritos)))
plt.subplots_adjust(hspace=0.5)

# Título
fig.suptitle('Distribución (%) de vehículos por distritos en función su Antigüedad: años 2019, 2020 y 2021', fontsize=16)

# Iteramos sobre cada 'Nom_Districte'
for i, Distrito in enumerate(Distritos):
    # Filtramos los datos por 'Nom_Districte'
    Distrito_df = Antigüedad_filtered_df[Antigüedad_filtered_df['Nom_Districte'] == Distrito]

    # Iteramos sobre cada 'Any'
    for j, año in enumerate([2019, 2020, 2021]):
        # Filtramos los datos por 'Any'
        año_df = Distrito_df[Distrito_df['Any'] == año]

        # Calculamos el total de 'Nombre' por 'Antiguitat'
        Total_por_distrito = año_df.groupby('Tipus_Vehicles')['Nombre'].sum()
           
        # Configuramos los datos y etiquetas para el gráfico circular
        labels = Total_por_distrito.index
        sizes = Total_por_distrito.values
          
        # Generamos el gráfico
        ax = axes[i, j]
        ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'{Distrito} - {año}')
        
# Leyenda
legend_handles, legend_labels = ax.get_legend_handles_labels()
fig.legend(legend_handles, legend_labels, title='Tipo vehículo', loc='upper right')

# Para el ajuste del espacio entre el título y los subplots
plt.tight_layout(rect=[0, 0.05, 1, 0.95])

plt.show()
```

Proporcion de vehículos en cada uno de los Distritos por __Antigüedad__ y __Año__:

```python
# Filtramos los datos por años (2019, 2020 y 2021):
Antigüedad_filtered_df = Antigüedad_tipo_TOTAL[Antigüedad_tipo_TOTAL['Any'].isin([2019, 2020, 2021])]

# Lista de valores únicos de 'Nom_Districte'
Distritos = Antigüedad_filtered_df['Nom_Districte'].unique()

# Tamaño y diseño de los subplots
fig, axes = plt.subplots(nrows=len(Distritos), ncols=3, figsize=(15, 5*len(Distritos)))
plt.subplots_adjust(hspace=0.5)

# Título
fig.suptitle('Distribución (%) de vehículos por distritos en función su Antigüedad: años 2019, 2020 y 2021', fontsize=16)

# Iteramos sobre cada 'Nom_Districte'
for i, Distrito in enumerate(Distritos):
    # Filtramos los datos por 'Nom_Districte'
    Distrito_df = Antigüedad_filtered_df[Antigüedad_filtered_df['Nom_Districte'] == Distrito]

    # Iteramos sobre cada 'Any'
    for j, año in enumerate([2019, 2020, 2021]):
        # Filtramos los datos por 'Any'
        año_df = Distrito_df[Distrito_df['Any'] == año]

        # Calculamos el total de 'Nombre' por 'Antiguitat'
        Total_por_distrito = año_df.groupby('Antiguitat')['Nombre'].sum()
           
        # Configuramos los datos y etiquetas para el gráfico circular
        labels = Total_por_distrito.index
        sizes = Total_por_distrito.values
          
        # Generamos el gráfico
        ax = axes[i, j]
        ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'{Distrito} - {año}')
        
# Leyenda
legend_handles, legend_labels = ax.get_legend_handles_labels()
fig.legend(legend_handles, legend_labels, title='Antigüedad', loc='upper right')

# Para el ajuste del espacio entre el título y los subplots
plt.tight_layout(rect=[0, 0.05, 1, 0.95])

plt.show()
```

Proporcion de vehículos en cada uno de los Distritos por __Distintivo Ambiental__ y __Año__:

```python
# Filtramos los datos por años (2019, 2020 y 2021):
Distintivo_AMB_filtered_df = Distintivo_AMB_TOTAL[Distintivo_AMB_TOTAL['Any'].isin([2019, 2020, 2021])]

# Lista de valores únicos de 'Nom_Districte'
Distritos_AMB = Distintivo_AMB_filtered_df['Nom_Districte'].unique()

# Tamaño y diseño de los subplots
fig, axes = plt.subplots(nrows=len(Distritos_AMB), ncols=3, figsize=(15, 5*len(Distritos_AMB)))
plt.subplots_adjust(hspace=0.5)

# Título
fig.suptitle('Distribución (%) de vehículos por distritos en función su Distintivo Ambiental: años 2019, 2020 y 2021', fontsize=16)

# Iteramos sobre cada 'Nom_Districte'
for i, Distrito_AMB in enumerate(Distritos_AMB):
    # Filtrar los datos por 'Nom_Districte'
    Distrito_AMB_df = Distintivo_AMB_filtered_df[Distintivo_AMB_filtered_df['Nom_Districte'] == Distrito_AMB]

    # Iteramos sobre cada 'Any'
    for j, año_AMB in enumerate([2019, 2020, 2021]):
        # Filtramos los datos por 'Any'
        año_AMB_df = Distrito_AMB_df[Distrito_AMB_df['Any'] == año_AMB]

        # Calculamos el total de 'Nombre' por 'Desc_Distintiu_Ambiental'
        Total_por_distrito_AMB = año_AMB_df.groupby('Desc_Distintiu_Ambiental')['Nombre'].sum()
           
        # Configuramos los datos y etiquetas para el gráfico circular
        labels = Total_por_distrito_AMB.index
        sizes = Total_por_distrito_AMB.values
        colors = ['blue', 'yellow', 'green', 'lightgreen', 'gray'] # En este caso personalizamos el color para que sea igual al de la etiqueta AMB
  
         # Generamos el gráfico
        ax = axes[i, j]
        ax.pie(sizes, labels=labels, colors = colors, autopct='%1.1f%%', startangle=90)
        ax.set_title(f'{Distrito_AMB} - {año_AMB}')
        
# Leyenda
legend_handles, legend_labels = ax.get_legend_handles_labels()
fig.legend(legend_handles, legend_labels, title='Distintivo Ambiental', loc='upper right')

# Para el ajuste del espacio entre el título y los subplots
plt.tight_layout(rect=[0, 0.05, 1, 0.95])

# Mostrar los gráficos
plt.show()
```

#### Comenatrios:

- Realziando la distribución de vehiculos por distritos y distintivo AMB observamos que en Ciutat Vella el porcentaje de vehículos de '0' emisiones (Etiqueta 0 - Zero emissions - Blava) es muy alto respecto al de otros distritos, situandose en un 14% (2021) cuando en los otros distritos están como máximo con un 2%.

### Análisis tipos de vehiculos '0' emisiones en Ciutat Vella

- Analizamos que tipos de vehículos de Ciutat Vella tienen mayor cantidad de etiqueta '0' emisiones (Etiqueta 0 - Zero emissions Blava)

```python
# Tratamos de visualitzar la cantidad por tipos de vehículos '0' emisiones por año en el Distrito de Ciutat Vella

    # Creamos un DF solo de Ciutat Vella y Desc_Distintiu_Ambiental = Etiqueta 0 - Zero emissions - Blava

df_CV_Cero_emisiones = Distintivo_AMB_TOTAL[(Distintivo_AMB_TOTAL['Nom_Districte'] == "Ciutat Vella") & (Distintivo_AMB_TOTAL['Desc_Distintiu_Ambiental'] == "Etiqueta 0 - Zero emissions - Blava")]

df_CV_Cero_emisiones

# Dado que queremos analizar por 'Tipues_vehicles' generamos un nuevo DF agrupando el total por tipo de vehiculo y año

df_CV_Cero_emisiones_agg = df_CV_Cero_emisiones.groupby(["Tipus_Vehicles", "Any"])["Nombre"].sum().reset_index()

# Hacemos el nuevo DF con la función "df.pivot" 

heatmap_df = df_CV_Cero_emisiones_agg.pivot("Tipus_Vehicles","Any","Nombre")

# Visualización del Heatmap

f, ax = plt.subplots(figsize=(10, 10)) # Per modificar mida del gràfic
sns.heatmap(heatmap_df, annot=True, ax=ax, fmt=",") # annot=True: per a que surtin els valors, fmt: per modificar visualització valor

plt.show()
```

#### Comenatrios:

- Al analizar por tipo de vehículo las etiquetas '0' emisiones, se observa que un número muy elevado respecto al total de vehiculos con etiqueta '0'en Ciutat Vella, corresponde a Ciclomotores. Coincide con el alto porcentaje de ciclomotores que tiene Ciutat Vella respecto al resto de Distritos.
- Probablemente se deba a que empresas de sharing de ciclomotores, esten censadas en Ciutat Vella: para el alquiler a turistas, etc.
